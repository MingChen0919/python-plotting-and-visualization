## Axes Appearance
---

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Create a figure with two subplots.

```python
fig, splots = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))
subplot_1 = splots[0]
subplot_1.axes.set_axisbelow(True)
subplot_1.axes.grid(b=True, which='major', axis='y', color='red', linestyle='-')
subplot_1.axes.set_facecolor('#D5F5E3')


subplot_2 = splots[1]
subplot_2.axes.set_axisbelow(False)
subplot_2.axes.grid(b=True, which='both', axis='x', color='blue', linestyle='-.')
subplot_2.axes.set_facecolor('#2ECC71')
plt.show()
```

## Axes legend
---

A **`legend`** instance represents a mapping between a graphic instance to annotation text. `Line2D` and `Patch` are graphic instances.

<span style="color:red">When we use the plotting methods from the **`Axes`** class to generate plots, the plotting functions not only draw plots, but also return graphic elements that we save to variables and then later use them in the **`legend()`** function to generate legends.</span>

```python

fig, splots = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

subplot_1 = splots[0]
subplot_2 = splots[1]

x = np.arange(0, 2, 0.01)
y1 = np.sin(2*np.pi*x)
y2 = np.exp(-x)

s = np.random.randn(200)
y3 = 2*s + 10 + np.random.randn(200)
y4 = s**2 + s + 10


## two lines in first subplot
l1, = subplot_1.plot(x, y1, marker='<', color='g')
l2, = subplot_1.plot(x, y2, marker='D', color='b')
subplot_1.axes.legend((l1, l2), ('l1', 'l2'))

## scatter points in second subplot
p1, = subplot_2.scatter(s, y3, marker = '+', color='#85C1E9')
# p2, = subplot_2.scatter(s, y4, marker = 'o', color='#CB4335')
subplot_2.axes.legend((p1, p2), ('p1', 'p2'))
plt.show()
```

### Why do we need commas when there is only one object?

In the following code, if we didn't add comma to the end of `l1` and `l2`, it would fail when we pass `l1` and `l2` to the **`legend()`** function.
