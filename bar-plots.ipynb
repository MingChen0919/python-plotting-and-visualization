```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Data

The position and size is determined by four parameters: **left and bottom** determine the position, and **width and height determine** the size. This is similar to how we specify the position and size of an **axes** instance. 

```python
n = 10
xleft = np.arange(n)
xbottom = 0
xwidth = 0.5
xheight = np.random.randn(n)*2 + 5


ybottom = np.arange(n)
yleft=0
yheight = 0.5
ywidth = np.random.randn(n)*2 + 5
```

## Bar plots

```python
import matplotlib.colors as mcolors
colors = list(mcolors.CSS4_COLORS.values())[-n:]

fig, splots = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
sp1 = splots[0]
sp2 = splots[1]

bars1 = sp1.bar(left=xleft, bottom=xbottom, width=xwidth, height=xheight, color=colors)
sp1.legend(bars1, colors)

bars2 = sp2.barh(left=yleft, bottom=ybottom, width=ywidth, height=yheight, color=colors)
sp2.legend(bars2, colors)

plt.show()
```
