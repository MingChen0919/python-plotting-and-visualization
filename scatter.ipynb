```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Data

```python
n = 20
x = np.arange(n)
y = np.random.randn(n)
```

## Scatter plot

```python
fig, splots = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

sp1 = splots[0]
sp2 = splots[1]

s1 = sp1.scatter(x, y, marker='|', color='red', )
s1.set_label('Scatter plot 1')
sp1.legend()
sp1.axes.set_xlabel('X')
sp1.axes.set_ylabel('Y')
sp1.set_title('Scatter plot 1')


s2 = sp2.scatter(x, y, marker='p', color='blue', s=x*10, edgecolors='r')
s2.set_label('Scatter plot 2')
sp2.legend()
sp2.axes.set_xlabel('X')
sp2.axes.set_ylabel('Y')
sp2.set_title('Scatter plot 2')

plt.show()
```
