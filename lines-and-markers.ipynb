# Lines and markers
---

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Data

```python
x = np.random.randn(50)
x.sort()
y1 = 2*x + np.random.randn(len(x))*5
y2 = 3*x + np.random.randn(len(x))*8
```

```python
fig, splots = plt.subplots(nrows=1, ncols=2, figsize=(12,5))
sp1 = splots[0]
sp2 = splots[1]

l1, = sp1.plot(x, y1, alpha=0.2, color='blue')
l2, = sp1.plot(x, y2, alpha=1, color='red')
l1.set_label('Line 1')
l2.set_label('Line 2')
sp1.legend()
sp1.axes.set_xlabel('X')
sp1.axes.set_ylabel('Y')
sp1.axes.set_title('Plotting lines')

m1, = sp2.plot(x, y1, alpha=0.2, marker='p', linestyle='None')
m2, = sp2.plot(x, y2, alpha=1, marker='D', linestyle='None')
m1.set_label('Marker 1')
m2.set_label('Marker 2')
sp2.legend()
sp2.axes.set_xlabel('X')
sp2.axes.set_ylabel('Y')
sp2.axes.set_title('Plotting markers')

plt.show()
```
